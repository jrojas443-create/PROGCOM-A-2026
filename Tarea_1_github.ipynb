```python

# 1. ECUACIÓN CUADRÁTICA
print("=== 1. ECUACIÓN CUADRÁTICA ===")
a = float(input("Ingrese a: "))
b = float(input("Ingrese b: "))
c = float(input("Ingrese c: "))

if a == 0:
    print("No es una ecuación cuadrática.")
else:
    d = b**2 - 4*a*c
    if d > 0:
        x1 = (-b + d**0.5) / (2*a)
        x2 = (-b - d**0.5) / (2*a)
        print("Dos soluciones:", x1, "y", x2)
    elif d == 0:
        print("Una solución:", -b/(2*a))
    else:
        print("No tiene soluciones reales.")
```

```python
# 2. TERCERA EDAD
print("\n=== 4. TERCERA EDAD ===")
nombre = input("Ingrese su nombre: ")
edad = int(input("Ingrese su edad: "))

if edad >= 70:
    print(nombre, "tiene prioridad en la fila.")
else:
    print(nombre, "no pertenece a la tercera edad.")
```

```python
# 3. MAYOR DE TRES NÚMEROS
print("\n=== 5. MAYOR DE TRES ===")
n1 = float(input("Número 1: "))
n2 = float(input("Número 2: "))
n3 = float(input("Número 3: "))

if n1 >= n2 and n1 >= n3:
    print("El mayor es:", n1)
elif n2 >= n1 and n2 >= n3:
    print("El mayor es:", n2)
else:
    print("El mayor es:", n3)
```

```python
# 4. AÑO BISIESTO
print("\n=== 3. AÑO BISIESTO ===")
anio = int(input("Ingrese un año: "))

if (anio % 4 == 0 and anio % 100 != 0) or (anio % 400 == 0):
    print("Es bisiesto.")
else:
    print("No es bisiesto.")
```

```python
# 5. PAR O IMPAR (TERNARIO)
print("\n=== 2. PAR O IMPAR ===")
num = int(input("Ingrese un número: "))
resultado = "Par" if num % 2 == 0 else "Impar"
print("Resultado:", resultado)
```

```python
# 6. TIPO DE TRIÁNGULO
print("\n=== 6. TIPO DE TRIÁNGULO ===")
l1 = float(input("Lado 1: "))
l2 = float(input("Lado 2: "))
l3 = float(input("Lado 3: "))

if l1 == l2 == l3:
    print("Equilátero")
elif l1 == l2 or l1 == l3 or l2 == l3:
    print("Isósceles")
else:
    print("Escaleno")
```

```python
# 7. CALCULADORA
print("\n=== 8. CALCULADORA ===")
x = float(input("Número 1: "))
y = float(input("Número 2: "))
op = input("Operador (+, -, *, /): ")

if op == "+":
    print("Resultado:", x + y)
elif op == "-":
    print("Resultado:", x - y)
elif op == "*":
    print("Resultado:", x * y)
elif op == "/":
    if y != 0:
        print("Resultado:", x / y)
    else:
        print("No se puede dividir por cero.")
else:
    print("Operador inválido.")
```

```python
# 8. IMC
print("\n=== 7. IMC ===")
nombre_imc = input("Nombre: ")
peso = float(input("Peso (kg): "))
altura = float(input("Altura (m): "))

imc = peso / (altura ** 2)

if imc < 18.5:
    print(nombre_imc, "- Bajo peso")
elif imc < 25:
    print(nombre_imc, "- Normal")
elif imc < 30:
    print(nombre_imc, "- Sobrepeso")
else:
    print(nombre_imc, "- Obesidad")
```

```python
# 9. CALIFICACIÓN
print("\n=== 9. CALIFICACIÓN ===")
nota = int(input("Ingrese la nota (0-100): "))

if nota >= 90:
    print("A")
elif nota >= 80:
    print("B")
elif nota >= 70:
    print("C")
elif nota >= 60:
    print("D")
else:
    print("F")
```

```python
# 10. VALIDAR TRIÁNGULO
print("\n=== 11. VALIDAR TRIÁNGULO ===")
t1 = float(input("Lado 1: "))
t2 = float(input("Lado 2: "))
t3 = float(input("Lado 3: "))

if t1 + t2 > t3 and t1 + t3 > t2 and t2 + t3 > t1:
    print("Forman triángulo")
    if t1 == t2 == t3:
        print("Equilátero")
    elif t1 == t2 or t1 == t3 or t2 == t3:
        print("Isósceles")
    else:
        print("Escaleno")
else:
    print("No forman triángulo")
```
