#BATALLA POR EQUIPOS
1. Calcular **quines** son los que sacaron mas que el promedio, y quien fue la mayor nota y con cuanto. considere que son 10 estudiantes.
2. Imprimir los votos obtenidos por 5 candidatos distintos. suponga que votan n usuarios.
seleccione los dos primeros y vuelva a ejecutar la votacion con esos dos(segunda vuelta).
3.Estana en la pantalla



```python
nombres = []
notas = []
for i in range(10):
    nombre = input("Nombre del estudiante: ")
    nota = float(input("Nota: "))
    nombres.append(nombre)
    notas.append(nota)
promedio = sum(notas) / len(notas)
print("Promedio:", promedio)
print("Estudiantes con nota mayor al promedio:")
for i in range(10):
    if notas[i] > promedio:
        print(nombres[i], notas[i])
mayor = max(notas)
pos = notas.index(mayor)
print("La mayor nota fue:", mayor)
print("La obtuvo:", nombres[pos])
```

```python
votos = [0,0,0,0,0]

n = int(input("Número de votantes: "))

# Primera vuelta
for i in range(n):
    voto = int(input("Vote por candidato (1-5): "))
    votos[voto-1] = votos[voto-1] + 1

print("Votos:", votos)

# Buscar los dos primeros
primero = votos.index(max(votos))

votos2 = votos.copy()
votos2[primero] = -1
segundo = votos2.index(max(votos2))

print("Pasan a segunda vuelta:", primero+1, "y", segundo+1)

# Segunda vuelta
final = [0,0]

for i in range(n):
    voto = int(input("Vote 1 o 2: "))
    final[voto-1] = final[voto-1] + 1

print("Resultado final:", final)

if final[0] > final[1]:
    print("Ganador candidato", primero+1)
else:
    print("Ganador candidato", segundo+1)
```
